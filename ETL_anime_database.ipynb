```python
import pandas as pd
import requests
import sqlite3 
```

```python
# --EXTRACT DATA--
url = 'https://anime-db.p.rapidapi.com/anime?page=1&size=10&search=Fullmetal&genres=Fantasy%2CDrama&sortBy=ranking&sortOrder=asc'
```

```python
header = {
    'X-RapidAPI-Key' : 'rapid_key',
    'X-RapidAPI-Host' : 'anime-db.p.rapidapi.com'
}
```

```python
response = requests.get(url, headers=header)
```

```python
data_json  = response.json()
```

```python
print(data_json)
```

```python
# -- CEK STRUKTUR KUNCI API SEBELUM DIUBAH KE TABLE --
print(data_json.keys())
```

```python
df = pd.DataFrame(data_json['data'])
```

```python
df.head()
```

```python
# --TRANSFORM--
df['genres'] = df['genres'].apply(lambda x: ','. join(x) if isinstance(x, list) else x)
df['alternativeTitles'] = df['alternativeTitles'].apply(lambda x: ','. join(x) if isinstance(x, list) else x)
```

```python
#-- LOAD (masukin ke sql) --
conn = sqlite3.connect('anime_database.db')
```

```python
df.to_sql('table_anime', conn, if_exists='replace', index=False)
```

```python
conn.close()
```
